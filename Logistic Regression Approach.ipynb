```python
import pandas as pd
import os
os.chdir('C:\\Users\\hp\\Desktop\\Python\\Kaggle\\Titanic Dataset')
import EDA_module as EDA
import sns_eda
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

disp=pd.options.display
disp.max_rows=100
disp.max_columns=100
```

```python
df=pd.read_csv('train.csv')
```

```python
EDA.EDA_summary(df)
```

```python
df
```

```python
#dropping the rows of data where there is even one null value
subset = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked', 'Survived']]
subset = subset.dropna().reset_index()
```

```python
#from the subset data extracting the essential columns as X and Y
X = pd.get_dummies(subset[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']], columns=['Sex', 'Embarked'])
Y = subset['Survived']
```

```python
#this creates the model of logistic regression
clf = LogisticRegression(fit_intercept=False).fit(X, Y)
```

```python
#this gives the model coefficients, but they are a bit different than the ones which statsmodel gives 
#so need to read and investigate more about the reason for the same
clf.coef_
```

```python
#this does a training prediction
clf.predict(X)
```

```python
#this gives me the percetage error
clf.score(X,Y)
```

```python
#this gives the confusion matrix 
#format goes something like [No, Yes] [No, Yes]
confusion_matrix(Y, clf.predict(X))
```

This shows that the out of 450 (365+85) people who didn't survive, the model incorrectly predicts 85 of them survived.
So Specificity = 85/450 = 18.8%

This shows that the out of 262 (59+203) people who did survive, the model incorrectly predicts 59 of them didn't survive.
So Sensitivity = 59/262 = 22.5%

Please note this is again for training data predictions
